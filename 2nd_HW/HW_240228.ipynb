## 실습4

```python
import pandas as pd
import numpy as np
from numpy.random import randn

np.random.seed(1234)
df = pd.DataFrame(randn(10, 4), columns=['A', 'B', 'C', 'D']) # 데이터 구성하기
df
```

```python
df['B'] # B열 선택
```

```python
df[['B', 'D']] # B,D 열 선택
```

```python
df[4:9] # 4 ~ 8행 선택
```

```python
df.loc[2:8,['A', 'C']] # 2 ~ 8행, A, C열 선택
```

```python
df.loc[:, ['A', 'C']] # A, C 열 선택
```

```python
df.loc[3:8] # 3 ~ 8행 선택
```

```python
df.loc[3:3] # 3행 단일 선택
```

```python
df.iloc[0:6, [0]] # 0 ~ 6행, A열 선택
```

```python
df.iloc[:, [0]] # A열 선택
```

```python
df.iloc[2:7] # 2 ~ 6행 선택
```

```python
df.iloc[5:6] # 5행 선택(단일행)
```

```python
df.query('D > 0.5') # D > 0.5 데이터 선택
```

```python
df.query('B > D') # B > D 데이터 선택 (그리고, slice 표현으로도 데이터 선택)
```

```python
df.query('A>0.5 & D<0.0') # A > 0.5 그리고 D < 0.0 데이터 선택
```

```python
df.query('A>0.5 | D<0.0') # A>0.5 혹은 D<0.0 데이터 선택
```

## 실습5

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
df2 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'],
                    'F': ['B4', 'B5', 'B6', 'B7'],
                    'G': ['C4', 'C5', 'C6', 'C7'],
                    'H': ['D4', 'D5', 'D6', 'D7']},
                    index=[0, 1, 2, 3])
df3 = pd.DataFrame({'I': ['A8', 'A9', 'A10', 'A11'],
                    'J': ['B8', 'B9', 'B10', 'B11'],
                    'K': ['C8', 'C9', 'C10', 'C11'],
                    'L': ['D8', 'D9', 'D10', 'D11']},
                    index=[0, 1, 2, 3])
```

```python
result = pd.concat([df1, df2, df3], axis=1)
result
```
