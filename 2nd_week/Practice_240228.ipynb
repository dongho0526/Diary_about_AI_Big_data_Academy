## 데이터 선택

```python
df = pd.DataFrame({'ABC': ['A', 'B', 'C', 'A'],
                   'abc': ['a', 'b', 'c', 'c'],
                   'group1':['G1', 'G2', 'G3', 'G4'],
                   'value1':[11, 13, 12, 14],
                   'val1_mod':[0.231, 0.197, 0.213, 0.199]})
df
```

```python
df.filter(like='a', axis=1) # 'a' 포함된 컬럼. regex='a'와 동일
df.filter(regex='[Aa]', axis=1) # 'A' 또는 'a' 포함된 컬럼([] 이용)
```

```python
df.mask(df.value1 > 12) # 조건 입력 및 mask 적용
df.mask(df.abc == 'a') # 조건 입력 및 mask 적용
df.mask(cond = df.value1 > 12, other = 999) # 조건에 해당되는 값 처리
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

```python
result = pd.concat([df1, df2, df3])
result
```

```python
result = pd.concat([df1, df2, df3], axis=1)
print(result)
result = pd.concat([df1, df2, df3], axis=1, ignore_index=True)
result.columns = ["col_" + str(i) for i in range(12)]
result
```
